```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
```

```python
data = pd.read_csv('/content/profile_pelanggan.csv')

data_kode = data['kode']
data_prov = data['provinsi']

unique_provinces = data_prov.unique()
num_provinces = len(unique_provinces)
colors = plt.cm.get_cmap('Set3', num_provinces)
color_mapping = {province: colors(i) for i, province in enumerate(unique_provinces)}

fig, ax = plt.subplots(figsize=(12, 6))

for province in unique_provinces:
    province_data = data[data['provinsi'] == province]
    ax.scatter(province_data['kode'], province_data['provinsi'], color=color_mapping[province], alpha=0.7, label=province)

ax.set_title('Scatter Plot - Kode vs. Provinsi')
ax.set_xlabel('Kode')
ax.set_ylabel('Provinsi')

ax.tick_params(axis='x', rotation=90)

ax.legend()

plt.show()

```

```python
data = pd.read_csv('/content/profile_pelanggan.csv')

data_kode = data['kode']
data_prov = data['provinsi']

unique_provinces = data_prov.unique()
num_provinces = len(unique_provinces)
colors = plt.cm.get_cmap('tab20', num_provinces)
color_mapping = {province: colors(i) for i, province in enumerate(unique_provinces)}

fig, ax = plt.subplots(figsize=(12, 6))

for province in unique_provinces:
    province_data = data[data['provinsi'] == province]
    ax.scatter(province_data['kode'], province_data['provinsi'], color=color_mapping[province], alpha=0.9, label=province)

ax.set_title('Scatter Plot - Kode vs. Provinsi')
ax.set_xlabel('Kode')
ax.set_ylabel('Provinsi')

ax.tick_params(axis='x', rotation=90)

ax.legend()

plt.show()
```
